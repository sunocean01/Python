## 查

### sections(): 提取文件中所有section

```python
sn = config.sections()
sn
```

### options('section'):获取指定section中的键,返回列表

```python
config.options('Plc')
```

### items():读取指定section中的键值对,返回形式:元组嵌套在列表中

```python
config.items('Plc')
```

### get('section','key'): 读取指定section下面指定的key的值

```python
config.get('Plc','netid')
```

### has_section()方法判断section是否存在
    存在返回True，不存在返回False

```python
config.has_section('login')
```

### has_option()方法判断section下，某个键是否存在
    存在返回True，不存在返回False

```python
config.has_option('Plc','netid')
```

## 增/改
    add_section():用来新增section
        add_section(section)
    set():用来新增对应section下的某个键值对
        set(section,key,value)

```python
# 在1.1中已经使用了read()方法读取了文件，所以在增加的时候都是增加到最后
# read()方法是用来读取配置文件的，如果不加上read()方法，写入是直接从头开始写的，使用了read()之后，是从读取完后的光标开始写入，类似追加模式'a'一样
```

```python
config.add_section('login')
config.set('login','id','123')
config.set('login','password','123')
```

```python
config.items('login')
```

```python
# 如果要修改login password值的话，重新set就可以
config.set('login','password','234')
```

## 删
    remove_section()方法删除某个section
        remove_section(section)
    remove_option()方法删除某个section下的键
        remove_option(section,key)

```python
config.remove_option('login','password')
```

```python
config.remove_section('login')
```

## 1. ini文件里的 option关键字被读取后,大写变成了小写

```python
config = cr.ConfigParser()   #实例化
inifile = r'.\testfile\FairfieldCalibration.ini'
ph = config.read(inifile)
config.items('CalibrationPreCheck')
print("注意: 'SkipFirst' 变成了 'skipfirst'")
```

```python
# 重写ConfigParser里面的 optionxform函数,取消小写转换, 之前是 return optionstr.lower()
class MyParser(ConfigParser):

    "Inherit from built-in class: ConfigParser"

    def optionxform(self,optionstr):

        "Rewrite without lower()"
        return optionstr
```

```python
config = MyParser()   #实例化
inifile = r'.\testfile\FairfieldCalibration.ini'
ph = config.read(inifile)
config.items('CalibrationPreCheck')
```
