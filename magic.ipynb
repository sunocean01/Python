# %%time: 代码执行时间

```python
%%time   #运行整个cell代码的时间

s = 0
for i in range(10):
    s = s + i
    time.sleep(0.5)
print(s)
    
```

```python
for i in range(3):
    %time time.sleep(0.5)      #运行单行代码的时间
```
