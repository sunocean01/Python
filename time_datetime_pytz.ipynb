## datetime.astimezone('目标时区'):将datetime附上时区信息

### 本地时区时间 --> UTC时间：datetime.astimezone()有效

```python
tgtz = pytz.timezone('UTC')   #目标时区
lotz = pytz.timezone('Asia/Shanghai')   #本地时区
dt = datetime.datetime.now(tz=lotz)  #必须标记这个时间是哪个时区的,否则会报错为naive time(不明确的一个时间);
dt.astimezone(tz=tgtz)
```

```python
dt
```

### UTC时间 --> 本地时区时间：datetime.astimezone() 无效

```python
tgz = pytz.timezone('Asia/Shanghai')   #目标时区
datetime.datetime.utcnow().astimezone(tz=tgz)
# 会发现结果时区名称是改了，但是时间没有改,是因为这个地方的datetime不是pytz里面的方法
```

### UTC时间 --> 本地时区时间: pytz.timezone(“本地时区”).fromutc("utc时间")

```python
# 基于上面的问题，需要使用下面方法：
tz = pytz.timezone('Asia/Shanghai')
dt = datetime.datetime.utcnow()
tz.fromutc(dt)
# .strftime("%Y-%m-%d %H:%M:%S")
```

### 其他时区间的时间转换
    1. pytz.timezone("当前时区").localize("datetime时间")                  :将时间附上时区信息
    2. pytz.timezone("当前时区").localize("datetime时间").astimezone("目标时区")   将带有时区的时间转换成目标时区时间

```python
tz = pytz.timezone("Asia/Shanghai")
tz.localize(datetime.datetime.today())   #datetime.datetime.today()本身是不带时区信息的
```

```python
# 将上面带有时区信息时间转换成其他时区时间
tz.localize(datetime.datetime.today()).astimezone(pytz.timezone("Africa/Juba"))
```

```python
tz.localize(datetime.datetime.today()).astimezone(pytz.timezone("UTC"))
```

# strftime():

```python
# %a 是星期的简写; 如 星期三:Web;
# %A 是星期的全写; 如 星期四: Wednesday;
# %b 月份的简写; 如 4月份: Apr;
# %B 月份的全写; 如 4月份:April;
# %c 日期时间的字符串表示; 如 04/07/10 10:43:39;
# %d 日在这个月中的天数(这个月的第几天);
# %f 微秒(范围[0,999999])
# %H 小时(24小时制,[0,23])
# %I 小时(12小时制,[0,23])
# %j 日在年中的天数[001,366](当年的第几天)
# %m 月份[01,12]
# %M 分钟[00,59]
# %p AM或者PM
# %S 秒[00,61], 为什么不是[00,59],参考python手册
# %U 周在当年的周数(是当年的第几周),星期天作为周的第一天;
# %W 周在当年的周数(是当年的第几周),星期一作为周的第一天;
# %w 今天在这周的天数[0,6],6表示星期天;
# %x 日期字符串(如: 04/07/10)
# %X 时间字符串(如: 10:43:39)
# %y 2个数字表示的年份,不带世纪;
# %Y 4个数字表示的年份,带世纪;
# %z 与utc时间的间隔(如果是本地时间,返回空字符串)
# %Z 时区名称(如果是本地时间,返回空字符串)
# %% %%-->%
```
