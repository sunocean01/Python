## 对全表进行函数应用

```python
# 该函数会将负数涂成红色，将正数涂成黑色
def color_negative_red(val):
    """
    Takes a scalar and returns a string with
    the css property `'color: red'` for negative
    strings, black otherwise.
    """
    color = 'red' if val > 0 else 'black'
    return f'color: {color}'
    # return "font-size:large"
```

```python
dt_style.style.applymap(color_negative_red,                  #applymap, 接收的是每一个元素,返回去的是CSS
                        subset=['FanSpeedStd_1','Omega'],    #subset, 可以给列标签列表,也可以给行标签列表
                       )  
```

```python
# dt_style.style.applymap(color_negative_red,subset=['2021-11-10 00:23:24','2021-11-10 00:28:03'])
```

```python
# 按条件指定背景色:
def highlight_max(s):
    '''
    highlight the maximum in a Series yellow.
    '''
    is_max = s == s.max()
    return ['background-color: yellow' if v else '' for v in is_max]
```

```python
dt_style.style.apply(highlight_max)
```

```python
# 链式操作,定义样式:
dt_style_d = dt_style[dt_style['Omega'].notna()]
dt_style_d.style.applymap(color_negative_red).apply(highlight_max)
```

```python
dt_style.style.applymap(lambda x: "background-color:red" if x>3 else '')
```

```python
dt_style.style.applymap(lambda x: "font-size:large" if x>3 else '')
```

# Finer control with slicing: 通过切片指定具体的列

```python
def highlight_max(s, props=''):
    return np.where(s == np.nanmax(s.values), props, '')
```

```python
slice_ = ['Omega']
```

```python
dt_style.style.apply(highlight_max, props='color:red;', axis=0, subset=slice_)\
.set_properties(**{'background-color': '#ffffb3'}, subset=slice_)
```
