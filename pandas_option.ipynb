```python
# pandas set_option的这些设置,只有有print函数时有效,jupyter自己的输出无效

pd.set_option("display.chop_threshold",11)  #设置阈值,小于该值的显示为 0

pd.set_option("display.colheader_justify","right")  #"left","right", 列标题靠哪边对其

pd.set_option("display.column_space", 12)    #列之间距离,默认12

pd.set_option("display.date_dayfirst",False)   #作用:当pandas要解析日期时,True:告诉pandas要解析的日期是 day first, 但这个不是严格的
pd.set_option("display.date_yearfirst",True)   # year first

pd.set_option("display.expand_frame_repr",True)   #是否自动换行,

pd.set_option("display.max_columns",10)   #显示最大列数
pd.set_option("display.max_rows",4)   #显示最大列数

pd.set_option("display.large_repr","truncate")   # 'inof','truncate': 打印时,是打印截取的内容,还是基本信息

pd.set_option("display.precision",2)   #显示小数点的位数

pd.set_option("display.show_dimensions",True)   #是否在最后显示DataFrame的行列信息

pd.set_option("max_colwidth",None)   #显示列中单独元素的最大长度

pd.get_option("display.show_dimension")  #获取对应的值
```
