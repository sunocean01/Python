```python
scale(item, xOrigin, yOrigin, xScale, yScale)
    — 缩放 item 指定的画布对象
    — xOrigin 和 yOrigin 决定要缩放的位置
    — xScale 和 yScale 决定缩放的比例
    — item 可以是单个画布对象的 ID，也可以是某个 Tag
    — 注意：该方法无法缩放 Text 画布对象

scan_dragto(x, y)
    — 见下方 scan_mark(x, y)

scan_mark(x, y)
    — 使用这种方式来实现 Canvas 内容的滚动
    — 需要将鼠标按钮事件及当前鼠标位置绑定到 scan_mark(x, y) 方法，然后再将 事件及当前鼠标位置绑定到 scan_dragto(x, y) 方法，就可以实现 Canvas 在当前位置和 sacn_mack(x, y) 指定的位置 (x, y) 之间滚动

select_adjust(item, index)
    – 调整选中范围，使得给定的 index 参数指定的位置在范围内
    – item 可以是单个画布对象的 ID，也可以是某个 Tag

select_clear()
    — 取消 Canvas 组件中所有选中的范围
    **

select_from(item, index)**
    — 调整选中范围的起始位置为 index 参数指定的位置
    — item 可以是单个画布对象的 ID，也可以是某个 Tag

select_item()
    — 范围在 Canvas 组件中当前文本的选中范围
    — 如果没有则返回 None

select_to(item, index)
    — 调整选中范围的结束位置为 index 参数指定的位置

tag_bind(item, event=None, callback, add=None)
    — 为 Canvas 组件上的画布对象绑定方法
    — event 参数是绑定的事件名称，callback 是与之关联的方法
    — item 可以是单个画布对象的 ID，也可以是某个 Tag
    — 注意：与绑定事件关联的是画布对象，而不是 Tag

tag_lower(item)
    — 将一个或多个画布对象移至底部
    — 如果是多个画布对象，将它们都移至底部并保留原有顺序
    — item 可以是单个画布对象的 ID，也可以是某个 Tag
    — 注意：该方法对窗口组件无效，请使用 lower 代替

tag_raise(item)
    — 将一个或多个画布对象移至顶部
    — 如果是多个画布对象，将它们都移至顶部并保留原有顺序
    — item 可以是单个画布对象的 ID，也可以是某个 Tag
    — 注意：该方法对窗口组件无效，请使用 lift 代替

tag_unbind(item, event, callback=None)
    — 解除与 item 绑定的事件
    — item 可以是单个画布对象的 ID，也可以是某个 Tag

tkraise(item, **options)
    — 将指定画布对象移动到显示列表的顶部
    — item 可以是单个画布对象的 ID，也可以是某个 Tag
    — 跟 tag_raise 一样

type(item)
    — 返回指定画布对象的类型
    — 返回值可以是：“arc”, “bitmap”, “image”, “line”, “oval”, “polygon”, “rectangle”, “text”, 或 “window”

xview(*args)
    — 该方法用于在水平方向上滚动 Canvas 组件的内容，一般通过绑定 Scollbar 组件的 command 选项来实现（具体操作参考：Scrollbar）
    — 如果第一个参数是 “moveto”，则第二个参数表示滚动到指定的位置：0.0 表示最左端，1.0 表示最右端
    — 如果第一个参数是 “scroll”，则第二个参数表示滚动的数量，第三个参数表示滚动的单位（可以是 “units” 或 “pages”），例如：xview(“scroll”, 3, “units”) 表示向右滚动三行

xview_moveto(fraction)
    — 跟 xview(“moveto”, fraction) 一样

xview_scroll(number, what)
    — 跟 xview(“scroll”, number, what) 一样

yview(*args)
    — 该方法用于在垂直方向上滚动 Canvas 组件的内容，一般通过绑定 Scollbar 组件的 command 选项来实现（具体操作参考：Scrollbar）
    — 如果第一个参数是 “moveto”，则第二个参数表示滚动到指定的位置：0.0 表示最顶端，1.0 表示最底端
    — 如果第一个参数是 “scroll”，则第二个参数表示滚动的数量，第三个参数表示滚动的单位（可以是 “units” 或 “pages”），例如：yview(“scroll”, 3, “pages”) 表示向下滚动三页

yview_moveto(fraction)
    — 跟 yview(“moveto”, fraction) 一样

yview_scroll(number, what)
    — 跟 yview(“scroll”, number, what) 一样
```

### 用鼠标画图：
    不得不说有点遗憾的是 Tkinter 并没有提供画“点”的方法。不过，我们可以通过绘制一个超小的椭圆形来表示一个“点”。下边例子中，通过响应“鼠标左键按住拖动”事件（），我们在鼠标拖动的同时获取鼠标的实时位置（x, y），并绘制一个超小的椭圆来代表一个“点”：

```python
root = tk.Tk()

w = tk.Canvas(root,width=400,height=200,background = "white")
w.pack()

def paint(event):
    x1,y1 = (event.x - 1),(event.y - 1)
    x2,y2 = (event.x + 1),(event.y + 1)
    w.create_oval(x1,y1,x2,y2,fill = "red")

w.bind("<B1-Motion>",paint)

tk.Label(root,text="按住鼠标左键并移动，开始绘制你理想蓝图吧.....").pack(side=tk.BOTTOM)

root.mainloop()
```

### onvalue/offvalue:修改variable的值

```python
# 默认情况下，variable 选项设置为 1 表示选中状态，反之设置为 0。你可以使用 onvalue 和 offvalue 选项修改它们的值，
# 例如下边代码，只要 var 被设置为“T”即选中状态，设置为“F”则相反：
window = tk.Tk(className="CheckbuttonWindow")
window.geometry("500x300")

v = tk.StringVar()
v.set("T")

c = tk.Checkbutton(master=window,text="脚踏实地",variable=v,onvalue="T",offvalue="F")
c.pack()

print(v.get())
window.mainloop()
```

```python
# 如果你需要同时跟踪 variable 选项以及 Checkbutton 组件对象，
# 你不妨可以试试将两者结合起来（下边在 CheckButton 组件对象中新建一个 var 变量，存放 variable 的值 v）：
# 默认情况下，variable 选项设置为 1 表示选中状态，反之设置为 0。你可以使用 onvalue 和 offvalue 选项修改它们的值，
# 例如下边代码，只要 var 被设置为“T”即选中状态，设置为“F”则相反：

window = tk.Tk(className="CheckbuttonWindow")
window.geometry("500x300")

v = tk.IntVar()
v.set(1)

c = tk.Checkbutton(master=window,text="脚踏实地",variable=v)

c.var = v   #这个是用户给创建的属性

c.pack()

print(c.var)

window.mainloop()
```

### 将代码放在类中，时时获取按钮状态

```python
# 如果你的 Tkinter 代码是放在类中的（在实际编程中你就应该这么干），那么将 variable 选项的值作为属性存储可能是更好的选择：

master = tk.Tk()


class A:
    
    def __init__(self, master):
        self.var = tk.IntVar()
        
        self.v = tk.StringVar()
        self.v.set("t")
        
        
        c = tk.Checkbutton(master, text="DUANG~", variable=self.var, command=self.cb)
        
        e = tk.Entry(master=master,textvariable=self.v,)
        
        c.pack()
        e.pack()
        
        master.mainloop()
     
    def cb(self):
        print("variable is", self.var.get())
        
        self.v.set(self.var.get())
    
a=A(master)
```

```python
window = tk.Tk(className="EntryWindow")
window.geometry("700x300")

tnames = round(random.random(),2)
cnames = tk.StringVar()
cnames.set(tnames)

    
e1 = tk.Entry(master=window,
             bg='white',   #或background,输入框背景颜色
             bd=4,         #或borderwidth,边框的大小,默认2个像素;
#              cursor='plus',#光标的形状设定，如arrow, circle, cross, plus 等
             font=('Arial',10),  #字体，型号，大小
              exportselection=0, #???默认情况下，你如果在输入框中选中文本，默认会复制到粘贴板，如果要忽略这个功能刻工艺设置 exportselection=0。
              fg='black',         #或foreground,文字颜色。值为颜色或为颜色代码，如：'red','#ff0000'
#               highlightbackground=,  #指定当 Entry 没有获得焦点的时候高亮边框的颜色,默认由系统指定
#               highlightcolor=,    #指定当 Entry 获得焦点的时候高亮边框的颜色，默认由系统指定
#               highlightthickness=, #指定高亮边框的宽度,默认值是 1 或 2 像素
#               insertbackground=,  #指定输入光标的颜色；如果被设置为非 0 值，光标样式会被设置为 RAISED；将 insertwidth 设置大一点才能看到效果哦
#               insertborderwidth=，#指定输入光标的边框宽度
              insertofftime=50,  #该选项控制光标的闪烁频率（灭）;单位是毫秒
              insertontime=50,   #该选项控制光标的闪烁频率（亮）;单位是毫秒
              insertwidth=5,    #指定光标的宽度;默认值是 1 或 2 像素
#               invalidcommand=,  #或invcmd,指定当输入框输入的内容“非法”时调用的函数;也就是指定当 validateCommand 选项指定的函数返回 False 时的函数
              
              justify='left',    #显示多行文本的时候,设置不同行之间的对齐方式，可选项包括LEFT, RIGHT, CENTER
              relief='flat',    #边框样式，设置控件3D效果，可选的有：FLAT、SUNKEN、RAISED、GROOVE、RIDGE。默认为 FLAT。
              selectbackground='blue',  #选中文字的背景颜色
              selectborderwidth=2,      #选中文字的背景边框宽度
              selectforeground='black',
#               show='-',    ##指定文本框内容显示为字符，值随意，满足字符即可。如密码可以将值设为 show="*"
              state='normal', #默认为 state=NORMAL, 文框状态，分为只读和可写，值为：normal/disabled
              takefocus=True,  #指定使用 Tab 键可以将焦点移动到输入框中;默认是开启的，可以将该选项设置为 False 避免焦点在此输入框中
              textvariable=cnames, #文本框的值，是一个StringVar()对象
#               validate=,    #该选项设置是否启用内容验证 ；见内容验证
#               validatecommand=,  #该选项指定一个验证函数，用于验证输入框内容是否合法,验证函数需要返回 True 或 False 表示验证结果,注意，该选项只有当 validate 的值非 "none" 时才有效
              width=30,     #文本框宽度,默认20
#               xscrollcommand=,设置水平方向滚动条，一般在用户输入的文本框内容宽度大于文本框显示的宽度时使用。
             
             )

e2 = tk.Entry(master=window)


e1.pack()
e2.pack(padx=20,
       pady=20,
       )
e2.insert("insert","Insert")   #向文本框中插入内容
e2.icursor(2)  #将指针移动到第二个字符后面,就是y的后面
# e2.selection_from(2)  #选中文本的开始点
# e2.selection_to(5)    #选中文本的结束点
# e2.selection_range(1,4)  #选中输入框的内容
print(e2.selection_present())  #bool,文本框的内容是否有被选中的
print(e2.get())    #获取文本框中的内容

window.mainloop()
```

```python
master = tk.Tk()
 
v = tk.StringVar()
 
def test(content, reason, name):
    if content == "CSDN":
        print("正确！")
        print(content, reason, name)
        return True
    else:
        print("错误！")
        print(content, reason, name)
        return False
 
testCMD = master.register(test)
e1 = tk.Entry(master, textvariable=v, validate="focusout", validatecommand=(testCMD, '%P', '%v', '%W'))
e2 = tk.Entry(master)
e1.pack(padx=10, pady=10)
e2.pack(padx=10, pady=10)
 
master.mainloop()
```

### 创建顶级菜单

```python
# 创建一个顶级菜单，你需要先创建一个菜单实例，然后使用 add() 方法将命令和其它子菜单添加进去：

window = tk.Tk(className="MenuWindow")
window.geometry("500x300")

def callback1():
    print("Hello")
def callback2():
    print("quit")

m = tk.Menu(master=window)
m.add_command(label="File",command=callback1)
m.add_command(label="quit",command=callback2)

window.config(menu=m)
window.mainloop()
```

### 创建下拉菜单

```python
window = tk.Tk(className="menuWindow")
window.geometry("500x300")

def callback1():
    print("Hello")
def callback2():
    print("quit")
def callback3():
    print("Cut")
def callback4():
    print("copy")
def callback5():
    print("Past")
def callback6():
    print("quit")    
    
    
    
    
mb = tk.Menu(master=window)

fm = tk.Menu(master=mb,
             tearoff=True,   #False: 下拉菜单头部没有虚线
            )
fm.add_command(label="open",command=callback1)
fm.add_command(label="save",command=callback2)
fm.add_separator()
fm.add_command(label="quit",command=window.quit)     #window.quit 会死机
mb.add_cascade(label="file",menu=fm)

em = tk.Menu(mb)
em.add_command(label="cut",command=callback3)
em.add_command(label="copy",command=callback4)
em.add_command(label="past",command=callback5)
mb.add_cascade(label="edit",menu=em)



window.config(menu=mb)

window.mainloop()
```

### post：弹出菜单

```python
# 在Frame区域，点击右键，会弹出窗口

window = tk.Tk(className="postwindow")
window.geometry("500x300")

def callback1():
    print("Yes")

mb = tk.Menu(master=window)
mb.add_command(label="Cancel",command=callback1)
mb.add_command(label="repeat",command=window.quit)

frame = tk.Frame(window,width=200,height=200,background="green")
frame.pack()

def popup(event):
    mb.post(event.x_root,event.y_root)

frame.bind("<Button-3>",popup)

window.config(menu=mb)

window.mainloop()
```

### 添加Checkbutton/Radiobutton

```python
window = tk.Tk(className="buttonWindow")
window.geometry("500x300")

def callback():
    print("Hello")

mb = tk.Menu(master=window)

openVar = tk.IntVar()
saveVar = tk.IntVar()
quitVar = tk.IntVar()

fm = tk.Menu(master=mb,tearoff=False)
fm.add_checkbutton(label="open",command=callback,variable=openVar)
fm.add_checkbutton(label="save",command=callback,variable=saveVar)
fm.add_separator()
fm.add_checkbutton(label="quit",command=window.quit,variable=quitVar)

mb.add_cascade(label="file",menu=fm)

editVar=tk.IntVar()

em = tk.Menu(master=mb,tearoff=False)
em.add_radiobutton(label="cut",command=callback,variable=editVar,value=1)
em.add_radiobutton(label="copy",command=callback,variable=editVar,value=2)
em.add_radiobutton(label="past",command=callback,variable=editVar,value=3)

mb.add_cascade(label="edit",menu=em)

window.config(menu=mb)

window.mainloop()
```

# Scale部件

## 参数

## 方法

## 用法

```python
# 创建一个指定范围的 Scale 组件其实非常容易，你只需要指定它的 from 和 to 两个选项即可。但由于 from 本身是 Python 的关键字，
# 所以为了区分需要在后边紧跟一个下划线：from_

window = tk.Tk(className="scall")
window.geometry("500x300")

def show():
    print(s1.get())
    print(s2.get())

s1 = tk.Scale(master=window,from_=0,to=42)
s1.pack()

s2 = tk.Scale(master=window,from_=0,to=100,orient="horizontal")
s2.pack()

tk.Button(master=window,text="get values",command=show).pack()


window.mainloop()
```

## tickinterval :设置刻度;length:设置长度(像素);resolution:步长

```python
window = tk.Tk(className="scall")
window.geometry("500x300")

s1 = tk.Scale(master=window,from_=0,to=42,tickinterval=10,length=150)
s1.pack()

s2 = tk.Scale(master=window,
              from_=0,
              to=100,
              orient="horizontal",
              tickinterval=20,
              length=400,
              resolution=10,     #resolution的值如果大于tickinterval，则interval就显示resolution的值
             )
s2.pack()


window.mainloop()
```

## 方法：

### insert：向文本框里插入内容

```python
window = tk.Tk(className="TextWindow")
window.geometry("500x300")

t = tk.Text(master=window)

# 在当前光标所在位置进行插入：
t.insert('insert','insert1\n')
t.insert(tk.INSERT,'insert2')

# 指定位置插入：
t.insert(2.1,"L")   #2.1: 第二行，位置2，这个位置和列表位置一样：0,1,2...

# 在最后插入：
t.insert('end','Y')
t.insert(tk.END,'K')

# 为插入的内容指定标签：
t.tag_config("tag_1", backgroun="yellow", foreground="red")  # 设置 tag
t.insert("end","puppy","tag_1")

t.pack()
window.mainloop()
```

### mark:将位置进行标记，跟踪
    解析：通俗的理解的就是看书时，碰到了重点，拿小本本来记录他在这边数的位置，如果再拿绿色或马克笔来画出来显眼一些这个就是tag标签
    默认是插入标记位置的右侧，可以设置gravity,插入左侧

#### Mark:用于表示位置

```python
window = tk.Tk(className="TextWindow")
window.geometry("500x300")

t=tk.Text(master=window,
      width=50,
      height=8)
t.pack()

t.insert('insert','dog is wangwang\n')
t.mark_set('here','1.7')   #要跟踪一个位置，那么你可以将该位置“标记”下来
t.insert("insert","cat is miaomiao")
t.insert('here','loud ')

window.mainloop()
```

### undo:“恢复”和“撤销操作”

```python
def back():
    t.edit_undo()

def forward():
    t.edit_redo()
```

```python
window = tk.Tk(className="TextWindow")
window.geometry("500x300")

t=tk.Text(master=window,
      width=50,
      height=8,
      undo=True
      )
t.insert('insert','dog is wangwang\n')
t.insert("insert","cat is miaomiao")

b1 = tk.Button(master=window,text="撤销",command=back)    #Text 的undo=True才有效
b2 = tk.Button(master=window,text="恢复",command=forward)

t.pack()
b1.pack()
b2.pack()

window.mainloop()
```

```python
def callback(envent):
    t.edit_separator() #人为地插入分割符
```

```python
window = tk.Tk(className="TextWindow")
window.geometry("500x300")

t=tk.Text(master=window,
      width=50,
      height=8,
      undo=True,
      autoseparators = False,   #如果要一个字符一个字符的撤销，此处需设置成False
      )
t.insert('insert','dog is wangwang\n')
t.insert("insert","cat is miaomiao")

#<Key>表示键盘事件，下面表示绑定了一个键盘事件，
#就是每次输入的时候就触发callback函数加一个分割符
t.bind("<Key>",callback)

b1 = tk.Button(master=window,text="撤销",command=back)    #Text 的undo=True才有效
b2 = tk.Button(master=window,text="恢复",command=forward)

t.pack()
b1.pack()
b2.pack()

window.mainloop()
```

### Indexes（索引）是用来指向 Text 组件中文本的位置
    跟 Python 的序列索引一样，Text 组件索引也是对应实际字符之间的位置。

#### 示例

```python
window = tk.Tk(className="TextWindow")
window.geometry("500x300")

t=tk.Text(master=window,
      width=50,
      height=8,
      undo=True,
      autoseparators = False,   #如果要一个字符一个字符的撤销，此处需设置成False
      )
t.insert('insert','dog is wangwang\n')
t.insert("insert","cat is miaomiao")

# 1. "line.column": "1.2"或1.2 都是可以的
print("'line.column:'",t.get("1.2",2.3),sep='\n')   #从1.2 到 2.3

# 2. "line.end":行号加上字符串 “.end” 的格式表示为该行最后一个字符的位置
print("'line.end:'",t.get("1.2","2.end"),sep='\n')

t.pack()

window.mainloop()
```

### tags:用于改变 Text 组件中内容的样式和功能

```python
window = tk.Tk(className="TextWindow")
window.geometry("500x300")

t=tk.Text(master=window,
      width=50,
      height=8,
      undo=True,
      autoseparators = False,   #如果要一个字符一个字符的撤销，此处需设置成False
      )
t.insert('insert','dog is wangwang\n')
t.insert("insert","cat is miaomiao\n")

# 1. 向现有内容应用标签
t.tag_add("tag1","1.7","1.12","1.14")
t.tag_config("tag1",background = "yellow",foreground = "red")

# 2. 对要插入的内容应用标签：
t.tag_config("tag2",foreground="yellow",background="green")   #更多参数见下面列表
t.insert("end","This is tag2","tag2")



t.pack()

window.mainloop()
```

####  tag_raise() ，tag_lower() ：提高或降低tag的优先级

```python
root = tk.Tk()

text = tk.Text(root, width = 30, heigh = 5)
text.pack()

text.tag_config("tag1", background="yellow", foreground="red")
text.tag_config("tag2", foreground="white")

#降低tag2的优先级 
text.tag_lower("tag2")

# 注意，与下边的调用顺序没有关系
text.insert(tk.INSERT, "I love Dogdog.com", ("tag2", "tag1"))

root.mainloop()
```

#### tag_bind():事件绑定

```python
window = tk.Tk()
window.geometry("500x300")

t = tk.Text(master=window,
           height=8,
           width=50,
           )
t.pack()

t.insert(tk.INSERT,"I love baidu.com")

t.tag_add("link","1.7","1.16")
t.tag_config("link",foreground="blue",underline=True)

def show_arrow_cursor(event):
    t.config(cursor="hand2")
    
def show_xterm_cursor(event):
    t.config(cursor="xterm")
    
def click(event):
    webbrowser.open("https://www.baidu.com")

t.tag_bind("link","<Enter>",show_arrow_cursor)
t.tag_bind("link","<Leave>",show_xterm_cursor)
t.tag_bind("link","<Button-1>",click)

window.mainloop()
```
